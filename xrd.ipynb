```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from pymatgen.analysis.diffraction.xrd import XRDCalculator
import tabulate 

with MPRester ("L8GeiTzK0JDCUblX") as m:
    
    s=m.get_structure_by_material_id("mp-2251")
    w=CifWriter(s)
    w.write_file('mp-2251')
    
    xrd = XRDCalculator(wavelength='CuKa',symprec=0,debye_waller_factors=None)
    
    xrdPattern = xrd.get_pattern(structure=s, scaled = True, two_theta_range = (0,90))
    

    print(xrdPattern.hkls)
    print("-----")

    
    for i in range (0,3):
        print("Pic défractogramme" , i+1)
        print("2 Théta : "  , xrdPattern.x[i])
        print("Intensité : ", xrdPattern.y[i])
        print(xrdPattern.hkls[i][0])
        print('\n')
        
        
    xrd.get_plot(s)
    
    
```
